```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
methods = ['NGCF', 'Mult-VAE', 'GRMF', 'GRMF-norm', 'LightGCN']
datasets = ['Gowalla', 'Yelp2018', 'Amazon-Book']

recall = [
    [0.1570, 0.0579, 0.0344],
    [0.1641, 0.0584, 0.0407],
    [0.1477, 0.0571, 0.0354],
    [0.1557, 0.0561, 0.0352],
    [0.1830, 0.0649, 0.0411],
]

ndcg = [
    [0.1327, 0.0477, 0.0263],
    [0.1335, 0.0450, 0.0315],
    [0.1205, 0.0462, 0.0270],
    [0.1261, 0.0454, 0.0269],
    [0.1554, 0.0530, 0.0315],
]

# Plotting
bar_width = 0.6
opacity = 0.8
index = np.arange(len(datasets))

fig, ax = plt.subplots()

for i, method in enumerate(methods):
    rects1 = ax.bar(index + i * bar_width / len(methods), recall[i], bar_width / len(methods), alpha=opacity, label=method)

ax.set_xlabel('Dataset')
ax.set_ylabel('Recall')
ax.set_title('Recall Comparison among LightGCN and Competing Methods')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(datasets)
ax.legend()

plt.show()


```
